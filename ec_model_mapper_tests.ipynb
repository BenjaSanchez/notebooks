## 1. Defining parser

```python
def parse_ec_model_ids(fluxes):
    import pandas as pd
    parsed_fluxes = pd.Series([])
    
    for rxn_id,flux in fluxes.iteritems():
        new_id = rxn_id
        new_flux = flux
        
        #Isozymes:
        if rxn_id.startswith("arm_"):
            new_id = rxn_id[4:]
        
        #Single enzyme:
        elif rxn_id.endswith("No1") and rxn_id[:-3] not in parsed_fluxes.index:
            new_id = rxn_id[:-3]
        
        parsed_fluxes.loc[new_id] = new_flux
        
        #Reversible rxns:
        if new_id.endswith("_REV") and flux > 0:
            new_id = new_id[:-4]
            new_flux = flux*-1
            parsed_fluxes.loc[new_id] = new_flux
        
        print("Before: " + rxn_id + " -> " + str(flux) + " / After: " + new_id + " -> " + str(new_flux))
        
    return parsed_fluxes
```

## 2. Testing with eciML1515

```python
import cobra
import pandas as pd
ecModel = cobra.io.read_sbml_model("eciML1515.xml")
solution = ecModel.optimize()
solution.fluxes
```

```python
parsed_fluxes = parse_ec_model_ids(solution.fluxes)
parsed_fluxes
```
